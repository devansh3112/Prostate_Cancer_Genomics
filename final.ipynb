### About Dataset

The final dataset provides a comprehensive resource for analyzing gene expression and genomic variations across different prostate cancer conditions.

**Data Sources**:

Initially, several GSE datasets with different GPL platforms were downloaded from the NCBI database. After an in-depth analysis, it was determined that GPL570, GPL96, and GPL571 platforms share significant overlap. Therefore, datasets corresponding to these three platforms were selected for further analysis. The raw data for each GSE was downloaded from the NCBI GEO database (https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi). If a dataset lacked raw data, it was excluded from the selection. The four selected GSE datasets are:
- GSE32448
- GSE46602
- GSE7307
- GSE69223

**Preprocessing**:  
The selected datasets were categorized into three groups based on the "Gillison" classification:
- Normal
- Benign
- Tumor  
Subsequently, the quality of the data was assessed using the MetaQC package in R. The results confirmed that these four datasets were of the highest quality and suitable for further analysis.

````python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report

# Step 1: Load the data with proper header alignment
df = pd.read_csv('Prostate_Cancer_Genomics.csv', header=1)

# Step 2: Extract Gillison classification from column names (after underscore)
# Ignore 'SYMBOL' column while processing the headers
sample_cols = [col for col in df.columns if col != 'SYMBOL']

# Extract classifications from headers using underscores
def extract_gillison_class(col_name):
    parts = col_name.split('_')  # Split by '_'
    return int(parts[-1])  # Assume the last part is the classification (1, 2, or 3)

# Step 3: Create a metadata DataFrame with GSM IDs and classifications
metadata = pd.DataFrame({
    'GSM_ID': sample_cols,
    'Gillison_Class': [extract_gillison_class(col) for col in sample_cols]
})

# Step 4: Map Gillison classes to Cancer/Non-Cancer
classification_map = {1: 'Non-Cancer', 2: 'Cancer', 3: 'Cancer'}
metadata['Cancer_Status'] = metadata['Gillison_Class'].map(classification_map)

# Step 5: Reshape the original data (gene vs. sample matrix)
df.set_index('SYMBOL', inplace=True)
df_transposed = df.T.reset_index().rename(columns={'index': 'GSM_ID'})

# Step 6: Merge metadata with transposed data
merged_data = pd.merge(metadata, df_transposed, on='GSM_ID')

# Step 7: Generate Confusion Matrix (Assume real vs. predicted labels for testing)
true_labels = merged_data['Cancer_Status']
predicted_labels = np.random.choice(['Non-Cancer', 'Cancer'], size=len(true_labels))  # Placeholder

# Step 8: Evaluate performance
conf_matrix = confusion_matrix(true_labels, predicted_labels, labels=['Non-Cancer', 'Cancer'])
class_report = classification_report(true_labels, predicted_labels, target_names=['Non-Cancer', 'Cancer'])

# Display results
print("Confusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)
````

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from imblearn.over_sampling import SMOTE  # Install via: pip install imblearn
from catboost import CatBoostClassifier, Pool

# Step 1: Load the data
df = pd.read_csv('Prostate_Cancer_Genomics.csv', header=1)

# Step 2: Extract Gillison classification from column names
sample_cols = [col for col in df.columns if col != 'SYMBOL']

def extract_gillison_class(col_name):
    return int(col_name.split('_')[-1])

metadata = pd.DataFrame({
    'GSM_ID': sample_cols,
    'Gillison_Class': [extract_gillison_class(col) for col in sample_cols]
})

# Step 3: Map Gillison classes to Cancer/Non-Cancer
classification_map = {1: 'Non-Cancer', 2: 'Cancer', 3: 'Cancer'}
metadata['Cancer_Status'] = metadata['Gillison_Class'].map(classification_map)

# Step 4: Reshape the gene expression data
df.set_index('SYMBOL', inplace=True)
df_transposed = df.T.reset_index().rename(columns={'index': 'GSM_ID'})

# Step 5: Merge metadata with transposed data
merged_data = pd.merge(metadata, df_transposed, on='GSM_ID')

# Step 6: Prepare features and labels
X = merged_data.drop(['GSM_ID', 'Gillison_Class', 'Cancer_Status'], axis=1)
y = merged_data['Cancer_Status'].map({'Non-Cancer': 0, 'Cancer': 1})

# Step 7: Feature selection using SelectKBest
k = 50  # Select top 50 features
selector = SelectKBest(score_func=f_classif, k=k)
X_selected = selector.fit_transform(X, y)
selected_features = selector.get_feature_names_out(input_features=X.columns)

# Step 8: Handle class imbalance with SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_selected, y)

# Step 9: Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.3, random_state=42
)

# Step 10: Initialize and train the CatBoost model
cat_model = CatBoostClassifier(
    iterations=1000,
    learning_rate=0.1,
    depth=6,
    eval_metric='Accuracy',
    random_seed=42,
    verbose=100
)

# Step 11: Cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(cat_model, X_train, y_train, cv=cv, scoring='accuracy')

# Visualize Cross-Validation Scores
plt.figure(figsize=(8, 5))
sns.boxplot(data=cv_scores, orient='h', palette='Set2')
plt.title('Cross-Validation Accuracy Distribution')
plt.xlabel('Accuracy')
plt.show()

print(f"Cross-Validation Accuracy: {np.mean(cv_scores):.2f}")

# Step 12: Train the final model on the training data
cat_model.fit(X_train, y_train, eval_set=(X_test, y_test), verbose=False)

# Extract feature importances
feature_importances = cat_model.get_feature_importance()
important_features = pd.Series(feature_importances, index=selected_features)

# Plot Feature Importances
plt.figure(figsize=(10, 6))
important_features.nlargest(10).plot(kind='barh', color='teal')
plt.title('Top 10 Feature Importances')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()

# Step 13: Make predictions
y_pred = cat_model.predict(X_test)

# Step 14: Evaluate the model
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, target_names=['Non-Cancer', 'Cancer'])
accuracy = accuracy_score(y_test, y_pred)

# Confusion Matrix Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=['Non-Cancer', 'Cancer'], yticklabels=['Non-Cancer', 'Cancer'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print(f"Test Accuracy: {accuracy:.2f}")
print("\nClassification Report:\n", class_report)

# Plot Training and Validation Accuracy over Iterations
plt.figure(figsize=(8, 5))
eval_results = cat_model.get_evals_result()
train_acc = eval_results['learn']['Accuracy']
val_acc = eval_results['validation']['Accuracy']

plt.plot(train_acc, label='Training Accuracy', color='green')
plt.plot(val_acc, label='Validation Accuracy', color='blue')
plt.title('Training vs. Validation Accuracy Over Iterations')
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
````

````python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping

# Step 1: Load the data
df = pd.read_csv('Prostate_Cancer_Genomics.csv', header=1)

# Step 2: Extract Gillison classification from column names
sample_cols = [col for col in df.columns if col != 'SYMBOL']

def extract_gillison_class(col_name):
    return int(col_name.split('_')[-1])

metadata = pd.DataFrame({
    'GSM_ID': sample_cols,
    'Gillison_Class': [extract_gillison_class(col) for col in sample_cols]
})

# Step 3: Map Gillison classes to Cancer/Non-Cancer
classification_map = {1: 'Non-Cancer', 2: 'Cancer', 3: 'Cancer'}
metadata['Cancer_Status'] = metadata['Gillison_Class'].map(classification_map)

# Step 4: Reshape the gene expression data
df.set_index('SYMBOL', inplace=True)
df_transposed = df.T.reset_index().rename(columns={'index': 'GSM_ID'})

# Step 5: Merge metadata with transposed data
merged_data = pd.merge(metadata, df_transposed, on='GSM_ID')

# Step 6: Prepare features and labels
X = merged_data.drop(['GSM_ID', 'Gillison_Class', 'Cancer_Status'], axis=1)
y = merged_data['Cancer_Status'].map({'Non-Cancer': 0, 'Cancer': 1})  # Encode labels

# Step 7: Feature selection using SelectKBest
k = 50  # Select top 50 features
selector = SelectKBest(score_func=f_classif, k=k)
X_selected = selector.fit_transform(X, y)

# Step 8: Handle class imbalance with SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_selected, y)

# Step 9: Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.3, random_state=42
)

# Step 10: Build the Sequential Neural Network
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))  # Input layer
model.add(BatchNormalization())
model.add(Dropout(0.3))  # Dropout to prevent overfitting

model.add(Dense(32, activation='relu'))  # Hidden layer
model.add(Dropout(0.3))

model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification

# Step 11: Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Step 12: Train the model with early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    X_train, y_train, validation_data=(X_test, y_test),
    epochs=100, batch_size=16, callbacks=[early_stopping], verbose=1
)

# Step 13: Evaluate the model
y_pred = (model.predict(X_test) > 0.5).astype("int32")

conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, target_names=['Non-Cancer', 'Cancer'])
accuracy = accuracy_score(y_test, y_pred)

# Display results
print(f"Test Accuracy: {accuracy:.2f}")
print("\nConfusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)

# Step 14: Plot Training History
plt.figure(figsize=(12, 4))

# Plot loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Over Epochs')
plt.legend()

# Plot accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy Over Epochs')
plt.legend()

plt.show()

# Additional Visualizations

# 1. Class Distribution Before and After SMOTE
plt.figure(figsize=(12, 5))

# Before SMOTE
plt.subplot(1, 2, 1)
sns.countplot(x=y, palette="Set1")
plt.title('Class Distribution Before SMOTE')
plt.xlabel('Cancer Status')
plt.ylabel('Count')

# After SMOTE
plt.subplot(1, 2, 2)
sns.countplot(x=y_resampled, palette="Set2")
plt.title('Class Distribution After SMOTE')
plt.xlabel('Cancer Status')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

# 2. Feature Importance Visualization (using scores from SelectKBest)
feature_scores = selector.scores_[selector.get_support()]  # Scores of selected features
selected_features = X.columns[selector.get_support()]  # Names of selected features

plt.figure(figsize=(10, 6))
sns.barplot(x=feature_scores, y=selected_features, palette="viridis")
plt.title('Top 50 Selected Feature Importance')
plt.xlabel('F-Score')
plt.ylabel('Gene Symbol')
plt.tight_layout()
plt.show()

# 3. Confusion Matrix Heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=['Non-Cancer', 'Cancer'], yticklabels=['Non-Cancer', 'Cancer'])
plt.title("Confusion Matrix Heatmap")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# 4. ROC Curve and AUC Score
y_pred_proba = model.predict(X_test)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, roc_curve, auc, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from lime import lime_tabular
import shap

# Assuming X_resampled and y_resampled are prepared datasets
X_train, X_val, y_train, y_val = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# SSA Parameters
population_size = 30
iterations = 50
c_bound = (0.1, 100)
gamma_bound = (0.0001, 1)

# Initialize population randomly within bounds
population = np.random.uniform(low=[c_bound[0], gamma_bound[0]],
                                high=[c_bound[1], gamma_bound[1]],
                                size=(population_size, 2))

# For plotting
best_fitness_progression = []

# SSA Optimization Loop
for iter in range(iterations):
    fitness = []
    for salp in population:
        C, gamma = salp
        model = SVC(C=C, gamma=gamma, kernel='rbf', probability=True, random_state=42)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_val)
        fitness.append(accuracy_score(y_val, y_pred))
    
    # Identify the best salp (leader)
    best_idx = np.argmax(fitness)
    leader = population[best_idx]
    best_fitness_progression.append(fitness[best_idx])
    
    # Update salp positions
    for i in range(population_size):
        if i == 0:  # Update leader's position
            population[i] = leader
        else:  # Follow the leader
            population[i] = 0.5 * (population[i] + leader)
        
        # Boundary check for C and gamma
        population[i][0] = np.clip(population[i][0], c_bound[0], c_bound[1])
        population[i][1] = np.clip(population[i][1], gamma_bound[0], gamma_bound[1])

# Final optimized parameters
best_C, best_gamma = leader
final_model = SVC(C=best_C, gamma=best_gamma, kernel='rbf', probability=True, random_state=42)
final_model.fit(X_train, y_train)

# Evaluation
y_final_pred = final_model.predict(X_val)
y_prob = final_model.predict_proba(X_val)[:, 1]  # Probabilities for the positive class
final_accuracy = accuracy_score(y_val, y_final_pred)

# Print results
print(f"Optimized C: {best_C}")
print(f"Optimized gamma: {best_gamma}")
print(f"Validation Accuracy with Optimized SVM: {final_accuracy:.2f}")

# ROC Curve
fpr, tpr, _ = roc_curve(y_val, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()

# Confusion Matrix
cm = confusion_matrix(y_val, y_final_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Non-Cancer', 'Cancer'])
disp.plot(cmap='Blues')
plt.title('Confusion Matrix of Final Model')
plt.show()

# SHAP Analysis
explainer = shap.KernelExplainer(final_model.predict_proba, shap.kmeans(X_train, 10))
shap_values = explainer.shap_values(X_val[:100])  # Analyze the first 100 validation samples

plt.figure()
shap.summary_plot(shap_values, X_val[:100], plot_type="bar")

# LIME Explanation for a single sample
lime_explainer = lime_tabular.LimeTabularExplainer(X_train.values, training_labels=y_train.values, mode='classification')
sample_idx = 0  # Replace with the index of the sample you want to explain
lime_exp = lime_explainer.explain_instance(X_val.values[sample_idx], final_model.predict_proba, num_features=10)

# Turn LIME explanation into an image
plt.figure(figsize=(8, 6))
lime_exp.as_pyplot_figure()
plt.title('LIME Explanation')
plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, roc_curve, auc, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.preprocessing import MinMaxScaler
from lime import lime_tabular
import shap
from imblearn.over_sampling import SMOTE

# Assume X_resampled and y_resampled are prepared
# 1. Feature scaling for feature selection
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_resampled)

# 2. Feature selection
k_best = 50  # Select top 50 features
selector = SelectKBest(mutual_info_classif, k=k_best)
X_selected = selector.fit_transform(X_scaled, y_resampled)

# 3. Train-test split
X_train, X_val, y_train, y_val = train_test_split(X_selected, y_resampled, test_size=0.3, random_state=42)

# 4. SMOTE for class balancing (if needed)
smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)

# Voting Classifier: Combine Logistic Regression, Decision Tree, and SVM
clf1 = LogisticRegression(random_state=42, max_iter=1000)
clf2 = DecisionTreeClassifier(random_state=42)
clf3 = SVC(C=1, gamma=0.1, kernel='rbf', probability=True, random_state=42)

voting_clf = VotingClassifier(
    estimators=[('lr', clf1), ('dt', clf2), ('svm', clf3)],
    voting='soft'
)

# 5. Fit the model
voting_clf.fit(X_train, y_train)

# 6. Predictions
y_pred = voting_clf.predict(X_val)
y_prob = voting_clf.predict_proba(X_val)[:, 1]  # Probabilities for the positive class

# Evaluation Metrics
accuracy = accuracy_score(y_val, y_pred)
print(f"Validation Accuracy with Voting Classifier: {accuracy:.2f}")

# ROC Curve
fpr, tpr, _ = roc_curve(y_val, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='green', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()

# Confusion Matrix
cm = confusion_matrix(y_val, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Non-Cancer', 'Cancer'])
disp.plot(cmap='Blues')
plt.title('Confusion Matrix of Voting Classifier')
plt.show()

# SHAP Analysis for Voting Classifier
explainer = shap.KernelExplainer(voting_clf.predict_proba, shap.kmeans(X_train, 10))
shap_values = explainer.shap_values(X_val[:100])  # Analyze the first 100 validation samples

plt.figure()
shap.summary_plot(shap_values, X_val[:100], plot_type="bar")

# LIME Explanation for a Single Sample
lime_explainer = lime_tabular.LimeTabularExplainer(X_train, training_labels=y_train, mode='classification')
sample_idx = 0  # Index of the sample to explain
lime_exp = lime_explainer.explain_instance(X_val[sample_idx], voting_clf.predict_proba, num_features=10)

# Display LIME Explanation as an Image
fig = lime_exp.as_pyplot_figure()
fig.set_size_inches(10, 8)  # Set figure size for better visualization
plt.tight_layout()  # Adjust layout to fit all elements
plt.show()

````
